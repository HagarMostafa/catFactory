```python
import numpy as np
```

```python
import csv
x = []
y = []
vx = []
vy = []
with open('ce889_dataCollectionN.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")
    for line in csv_reader:
        x.append(float(line[0]))
        y.append(float(line[1]))
        vx.append(float(line[2]))
        vy.append(float(line[3]))
```

```python
vx
```

```python
import random
w1 = random.random()
w2 = random.random()
bias = random.random()
lr = 0.05 #learning rate
print(w1,w2)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-1*x))
```

```python
def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
import numpy as np
for epoch in range(25000):
    XW = np.dot(x, w1)+ bias
    YW = np.dot(y, w2)+ bias
    av1 = sigmoid(XW)
    av2 = sigmoid(YW)
    error1 = av1 - vx
    error2 = av2 - vy
    print(error1.sum())
    print(error2.sum())
    dcost1 = error1
    dpred1 = sigmoid_derivative(vx)
    vx_del = dcost1 * dpred1
    xt = x.T
    w1 = w1 - lr*np.dot(xt, vx_del)
    
    for num in vx_del:
        bias = bias - lr*num
```
